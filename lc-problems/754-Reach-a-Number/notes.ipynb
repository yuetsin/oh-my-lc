# 754 Reach a Number

有趣的一道题。虽然看起来很像是普通的 DFS 可以解决的问题。

然而，考虑到数轴是无限的。假如我们真的秉持着「每个位置既可以向前走、又可以向后走」，那么绝对会在极短的时间里爆炸。
因为循环（几乎）没有终止的条件。

但是我们可以观察到，事实上移动的步数是存在上界的。假如我们每次都采取「进一步+退一步」的策略，就能实现花费 2 步走，实现数轴上的绝对移动 1 单位。

因此再怎么遥远的地方，都能在 2 * target 步以内到达！

遗憾的是，target 的范围在 [-10 ** 9, 10 ** 9] 以内。因此这个上界还是超级宽泛，根本没用。

最终我放弃了用编程的方法解决这个问题…因为这到底是个数学问题。

抽象成在 [1, k] 之间添加 + 、− 号使得结果为 target，同时要使得 k 最小。考虑到我们只要把所有的符号都反过来就能抵达相反数的地方，所以直接取绝对值运算。负号位没有意义。

然后，我们一直往正方向走，直到走到的结果已经超过或抵达了 k。假如刚刚好跳到这个 k，那么 lucky。

假如我们的 offset 是 2 的倍数，非常好，我们根本不需要更多的步数。靠把之前的某一步从 + 改成 -，我们可以让我们的落点往回走任意偶数单位距离。当然，是有界的——不会超过 k。

但如果我们的 offset 不是 2 的倍数，麻烦来了——我们调整之前的步数从 + 改成 -，都只能把结果以 2 的倍数移动。为了拿到这个奇数，我们必须再走几步。

走几步？假如 k 当前是偶数，那么下一步 k + 1 就是奇数，我们直接可以拿它来用；但如果非常非常不巧，k 是奇数，k + 1 是偶数，那么我们就得走到 k + 2 才能得到下一个奇数来破局。

总归，就是这样。
