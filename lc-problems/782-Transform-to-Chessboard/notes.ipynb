# 782 Transform to Chessboard

目的是：把我们的棋盘格给转换成「0、1 棋盘格交错」的格式。但是能做的只有交换任意行/任意列。

最後的「棋盘交错格」一定是「101010.....1」和「010101......0」交错的结果。视棋盘大小而定，也可能是「1010......0」和「0101......1」交错。

所以，实际上对于奇数长度的 N，「1」的个数一定是 N * N // 2 或者 N * N // 2 + 1。对于偶数长度的 N，则一定是 N * N // 2 个。

因此，先判断 1 的个数就能排除掉一些不可能的棋盘。

另外，提出一个议论：每一行/列都只能和第一行/列相同，或相补。

考虑到，进行任何「交换」，结果都必定仍然符合上面的条件：如果存在不符合的情况，那么「交换行列」也无法改变这种情况。

另外，只要每一行都和第一行相等或互补，那么列也就自然也满足「和第一列相等或互补」。

那么，只要满足 1）「1」的个数符合条件，且 2）每一行/列都和第一行/列相等或互补，那么这个棋盘格一定能够被归化。

在满足条件之後，我们只需要考虑交换行/列，使得每个「和第一行相同」的行和「和第一行互补」的行交替出现，就好了。

记录「相等」和「互补」两种不同的「列」为 P 和 N。那么，可以选择其归化到 PNPN…… 或者是 NPNP……。

记录出有多少列不在它该在的位置上。这个数字除以 2，就是需要交换的次数。从两种情况中挑选次数较少的那一种。得到结果。

就是那样。在完成行调换之後，再进行列调换，加起来就是结果。

关键是，要看出「有可能进行规划」的条件其实是很严格的，因为整个地调换行列其实不能很大地改变形态。
