如果要正经地计算的话，可以用这个性质：

一个偶完全数一定能被表示成这个形式：2^(n-1) * (2^n - 1)，其中 (2^n - 1) 是质数。也就是说，知道了一个梅森素数（表示为 2^n - 1 的质数），就能得到一个完全数。

这是充分必要条件，充分性由歐幾里得證明，而必要性則由歐拉所證明。

令 n = 2, 3, 5, 7, 13 就可以得到（在范围内的）前五个完全数了。留意到. n = 11 不可用，因为 2^11 - 1 = 2047 = 23 * 89 不是质数。
