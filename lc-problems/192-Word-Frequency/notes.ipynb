让写 Bash 脚本了…

cat words.txt | tr -s ' ' '\n' | sort | uniq -c | sort -r | awk '{ print $2, $1 }'

我们来慢慢分析分析这个…

cat words.txt 读取了文本内容，然后 | 交给 tr
tr -s 从这串字符中清除掉连续的 ‘ ’ 和 '\n' 两个字符，但仅仅保留其中一个以示分割，并交给 sort。
sort 会自动把上面这串字符串用 ' ' / '\n' 拆分成一个个的字符，并将其进行排序（确保连续单词的保留在一起）
uniq -c 会清除连续出现的重复行，并且计数（-c means counting）
sort -r 再次排序，不过这次是对计数之后的数字排序。因此我们用了 -r 来降序排序。
最后按照格式打印出来。

sort -r: -r means sorting in descending order
