因为写错了。没看到条件——去的时候只能往右往下走；回来的时候只能往左往上走。

这题的难处在于：如果我们第一次走到右下的时候，贪婪地拿走了最多的草莓，并且在走回左上的时候，也贪婪地拿走最多的草莓，这样两次贪婪的结果并不是最终的最大值。

理性的思考可以让我们理解：两条路所能采集到的草莓是相互关联的。有些草莓只能去的时候采到；有的只能在回来的路上采到。只有相互配合，才能实现最大值。

但是这题存在「历史性」，也就是不能简单地存储每个格点的 Memo；况且还有「回头」的设定，并不是很好套用原来的 DP 或是 DFS。

我们换个思路：把「回程」这条路给翻转过来。并不认为是一个人走了一去一回，而是认为有两个人同时从左上往右下走，最後在 (xmax - 1, ymax - 1) 处碰面。

考虑到因为我们在 x 和 y 方向上都不允许走回头路（去的路上只能右下、回来的路上只能左上），因此实际上我们每一程行走的步数是确定的——都是 xlen + ylen - 2。因此在每一个格点上，还能走的步数都已经确定了。

在同一时刻，上面的两个人所要走的步数都是一致的。设他们的坐标位置分别为 （x1, y1）和（x2, y2）。根据上面的步数关系，可以知道 x1 + y1 = x2 + y2。

这样，实际上我们只要三个坐标就能表示这两个位置啦。

那么在每个 (x1, y1, x2, y2) （这里虽然写了四个坐标，但我们记住只需要三个量就能表示。）关系处，我们有四种选择：

* Leg 1、2 都往右走
* Leg 1、2 都往下走
* Leg 1 往右走、Leg 2 往下走
* Leg 1 往下走、Leg 2 往右走

这样，我们就有了一个在 O(N^3) DP 下的动态规划问题。

留意到步数关系是很重要的；这直接把 N^4 压缩到了 N^3。
