# 777 Swim in Rising Water

又是 awice…猜谜式理解题意。

总归，你在一个布满石头的神秘海域里游泳。可以等待 1 个时间段，水就会上涨一格；当水涨到某一个石头的高度的时候（或者更高），就可以到这块石头上去。在石头之间游泳的时间忽略不计。问要多久才能从左上角游到右下角？

注意…awice 挖了个坑。`grid[0][0]` 并不保证是 0。在那不是 0 的时候，假如是 `t`，就必须要等待 `t` 时间单位之後才能下水开始游泳。

然后，说说解法。这里用 DFS 并不是很合适，因为短路条件不充分——理论上不论你怎么走，都能等足够长的时间之後游到终点——这就导致整棵树枝叶繁茂，非常费时。何况这里的 `N` 最大可能是 50。复杂度吃不消。
事实上，在提交的时候还没撞到 TLE，自己的递归次数就爆炸了…

这里，用并查集（Union-Find）的方式更适合。实现方式倒不要紧，我们的元素最多 2500 个，这里的性能不是瓶颈。

首先，最大的一个不同就是：我们不要游来游去了。反过来，我们只考察「连通性」。因为。你看，游泳根本不耗时间，而且实际上我们边游边等和一直等在起点最後一气呵成游过去也是一样的。

所以，重大的一个想法是，只需要考察水在上涨的过程中，何时起点和终点联通，就行了。并查集做这件事正拿手。

就算不用并查集合，写起来也不是很麻烦：不过注意不能简单的每次 + 1；对于这种「判定区域有二分边界值」的情况，我们直接用二分法找到 `t` 即可。
