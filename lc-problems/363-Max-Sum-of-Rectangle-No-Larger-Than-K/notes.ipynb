我真的写得这么烂吗。

这个复杂度是 O(n ^ 4)，因为需要反复构造四边形并计算。

首先 PreSum 这件事是没做错的；可以把每次的求和计算由 O(n) 压到 O(1)。

只是构造四边形要有点技巧：

提前把这个四边形沿着横边排序。

如此，即可利用「二分查找」来节约时间。

最终的时间复杂度可以压缩到 O(R*R*C*logC*logC)。

这也就是题目中的附加部分「假如这个 Matrix 非常扁，怎么办？（What if the number of rows is much larger than the number of columns?）」的意义。

因为在很扁的情况下，时间复杂度可能比较炸裂。

回答也是很简单的；做个转制。确保我们沿着较长的那条边进行 Sort／Binary Search。就好了。
